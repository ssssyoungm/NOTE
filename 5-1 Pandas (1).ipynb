#### Pandas Join

```python
import pandas as pd

technologies = {
    'Courses':["Spark","PySpark","Python","pandas"],
    'Fee' :[20000,25000,22000,30000],
    'Duration':['30days','40days','35days','50days'],
              }
index_labels=['r1','r2','r3','r4']
df1 = pd.DataFrame(technologies,index=index_labels)
print(df1)

technologies2 = {
    'Courses':["Spark","Java","Python","Go"],
    'Discount':[2000,2300,1200,2000]
              }
index_labels2=['r1','r6','r3','r5']
df2 = pd.DataFrame(technologies2,index=index_labels2)
print(df2)
```

```python
# pandas join
df3=df1.join(df2, lsuffix="_left", rsuffix="_right")
print(df3)
```

```python
# pandas Inner join DataFrames
df3=df1.join(df2, lsuffix="_left", rsuffix="_right", how='inner')
print(df3)
```

```python
# pandas Right join DataFrames
df3=df1.join(df2, lsuffix="_left", rsuffix="_right", how='right')
print(df3)
```

```python
# pandas join on columns
df3=df1.set_index('Courses').join(df2.set_index('Courses'), how='inner')
print(df3)
```

```python
# pandas join
df3=df1.join(df2.set_index('Courses'), how='inner', on='Courses')
print(df3)
```

#### Pandas Concat

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)
print(df1)

df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)
print(df2)

df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[8, 9, 10, 11],
)
print(df3)

frames = [df1, df2, df3]
result = pd.concat(frames)
print(result)
```

```python
result = pd.concat(frames, keys=["x", "y", "z"])
print(result)
```

```python
air_quality["측정일시"] = pd.to_datetime(air_quality["측정일시"], format="%Y%m%d")
```

```python
air_quality.head()
```

```python
air_quality["측정일시"].min(), air_quality["측정일시"].max()
```

```python
air_quality["측정일시"].max() - air_quality["측정일시"].min()
```

```python
air_quality.head()
```

```python
air_quality.groupby([air_quality['측정일시'].dt.weekday, "측정소명"])["이산화질소농도(ppm)"].mean()
```

```python
fig, axs = plt.subplots(figsize=(12, 4))

air_quality.groupby(air_quality['측정일시'].dt.weekday)["이산화질소농도(ppm)"].mean().plot(kind='bar', rot=0, ax=axs)

plt.xlabel("day")
plt.ylabel("ppm")
```
